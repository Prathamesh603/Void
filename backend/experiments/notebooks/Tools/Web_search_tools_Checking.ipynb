### Tavily search needs the ```TAVILY_API_KEY```

````python
from dotenv import load_dotenv
load_dotenv()
from langchain_tavily import TavilySearch
tool = TavilySearch(
    max_result = 5,
    topic='general'
)
result = tool.run("Todays gold price")
print("Type os the resutl : ",type(result))
print("Len of the rsult : ",len(result))
result
````

````python
print("1. The response contains this things")
print("2. Query : ",result.get("query"))
print("Results : ",result.get("results"))
print("Total results found : ",len(result.get("results")))
print("First Result : ",result.get("results")[0])
print("First Result URL : ",result.get("results")[0].get('url'))
print("First Result Title : ",result.get("results")[0].get('title'))
print("First Result Content : ",result.get("results")[0].get('content'))
print("First Result Score : ",result.get("results")[0].get('score'))
````

````python
all_results = result.get("results")
````

````python
all_results.sort(key=lambda x: x.get("score"),reverse=True)
for response in all_results :
    print("Title : ",response.get("title"),"Score : ",response.get("score"))
print("\n")
print("Best scores content : ",all_results[0].get("content"))
````

````python
from serpapi import GoogleSearch
import os
SERPAPI_API_KEY = os.getenv('SERPAPI_API_KEY')
def google_web_search(query):
    params = {
        "q": query,
        "engine": "google",
        "api_key": SERPAPI_API_KEY,
        "num": "2"
    }

    search = GoogleSearch(params)
    results = search.get_dict()

    return results
````

````python
response = google_web_search('LLMs')
````
