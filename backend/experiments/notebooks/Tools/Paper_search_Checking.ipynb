````python
from langchain_community.tools import ArxivQueryRun
from langchain_community.utilities import ArxivAPIWrapper

# Wrapper
arxiv = ArxivAPIWrapper(
    top_k_results=2,
)

# Tool
arxiv_tool = ArxivQueryRun(api_wrapper=arxiv)

# Direct tool call (without agent)
result = arxiv_tool.run("LLM's")

print(result)
````

### Here we are getting only abstract not link and all those thing becasue the {```ArxivQueryRun```} only returns only returns a formatted text summary (mostly the abstract).That tool is designed for quick Q&A style retrieval, not full metadata extraction.

### For full meta data we need to use the {```ArxivAPIWrapper.load()```}

````python
from langchain_community.utilities import ArxivAPIWrapper

arxiv = ArxivAPIWrapper(
    top_k_results=1,
    doc_content_chars_max=3000
)

docs = arxiv.load("LLMs")

doc = docs[0]

print("Title:", doc.metadata.get("Title"))
print("Authors:", doc.metadata.get("Authors"))
print("Published:", doc.metadata.get("Published"))
print("PDF:", doc.metadata.get("pdf_url"))
print("Entry ID:", doc.metadata.get("entry_id"))
print("Content",doc.metadata.get("Summary"))
````

````python
import arxiv

search = arxiv.Search(
    query="deep neural ntw",
    max_results=1
)

client = arxiv.Client()

for result in client.results(search):

    print("Title:", result.title)
    print("Published:", result.published)
    print("Authors:", result.authors)
    print("Abstract:", result.summary)
    print("PDF:", result.pdf_url)
    print("Abstract Page:", result.entry_id)
````

````python
import arxiv

client = arxiv.Client(
    page_size=5,
    delay_seconds=3,
    num_retries=5
)

search = arxiv.Search(
    query="transformers",
    max_results=1,
    sort_by=arxiv.SortCriterion.Relevance
)

for result in client.results(search):
    print("Title:", result.title)
````
