```python
import requests
import os

index = 1

def download_pdf(url,title):
    global index

    folder_path = "data/PDF"
    os.makedirs(folder_path, exist_ok=True)  # Create folders if not exist

    response = requests.get(url)

    file_path = os.path.join(folder_path, f"{title}.pdf")
    print(file_path)
    with open(file_path, "wb") as f:
        f.write(response.content)

    print(f"Saved: {file_path}")

    index += 1
```

```python
import fitz  # PyMuPDF

def extract_text(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text
```
